# 과제 1: 데이터 알아보기

## 1-1
AB_NYC_2019.csv 데이터 파일을 데이터 프레임 형태로 로드하여 df 변수에 저장합니다.

```python
import pandas as pd
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

## 1-2
- 로드한 데이터 일부를 살펴보고, 각 컬럼별 통계값을 뽑습니다.
- 결측치가 있다면 적절한 방법으로 처리합니다.

```python
df.sample(5)
```

```python
df.isna().sum(axis=0)
```

```python
# name 이 NaN 인 레코드를 확인한다. 숙소 정보에 대한 컬럼은 유효하여 레코드를 삭제하지 않는다.
df[df["name"].isna()]
```

```python
# host_name 이 NaN 인 레코드를 확인한다. 숙소 정보에 대한 컬럼은 유효하여 레코드를 삭제하지 않는다.
df[df["host_name"].isna()]
```

```python
# 결측치 컬럼 채우기
# name, host_name 은 임의의 값으로 채운다.
# last review 는 무시한다.
# review_per_month 는 평균값으로 채운다.
df["name"] = df["name"].fillna("NAME")
df["host_name"] = df["name"].fillna("HOST_NAME")
df["reviews_per_month"] = df["reviews_per_month"].fillna(df["reviews_per_month"].mean())
```

## 1-3
- 가격 (price) 컬럼에 대한 분포 파악을 위해, 기술 통계값을 뽑고 및 시각화 도구를 사용하여 그림을 그립니다.

```python
df["price"].describe()
```

```python
import seaborn as sns

sns.histplot(df["price"], kde=True)
```

```python
sns.violinplot(y=df["price"])
```

## 1-4
- 가격과 리뷰 수(number of reviews) 간의 관계를 살펴보기 위해, 관련된 기술 통계값을 뽑고 시각화 도구를 사용하여 그림을 그립니다.

```python
df[["price", "number_of_reviews"]].corr()

# 변수간 관련성이 관측되지 않는다.
```

```python
# 치우친 분포인 것을 고려하여 log 변환후 상관관계를 보자
import numpy as np 
df["log_price"] = np.log(df["price"] + 1)
df[["log_price", "number_of_reviews"]].corr()

# 변수간 관련성이 관측되지 않는다.
```

## 1-5
- neighborhood group과 가격의 관계를 알아보기 위해, 그룹별 가격의 분포에 대한 기술 통계값을 뽑고 시각화 도구를 사용하여 그림을 그려봅니다.

```python
df['neighbourhood_group'].unique()
```

```python
df.groupby('neighbourhood_group')[["price"]].agg(["mean", "median", "std"])

# 지역별 가격 분포의 차이가 명확히 나타나는 것을 알 수 있다.
```

```python
ax = sns.boxplot(data=df, x="neighbourhood_group", y="price")
ax.set_ylim(0, 300)

# 맨하탄 > 브루클린 순으로 가격이 높고, 나머지 지역은 비슷하다.
# 브롱크스가 조금 더 낮은 편
```

## 1-6: 그룹별 특징 자유 분석

- neighborhood group 에 따라 어떤 특징이 있는지 자유롭게 분석하고, 결과의 의미를 제시합니다.

지역에 따라 room type 별 가격 차이를 살펴보자.

```python
df.groupby(["neighbourhood_group", "room_type"])["price"].agg("mean", "median", "std")
```

룸 타입에 따라 가격이 영향을 많이 받는 것을 알 수 있으며 (entire home > private room > shared room), 지역별 가격 차이는 통합하여 분석하였을 떄와 동일하게 관측된다.

```python
sns.barplot(data=df, x="neighbourhood_group", y="price", hue="room_type")
```
