```python
import seaborn as sns


titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

print('승객 수:', len(df))
```

```python
df_grouped = df.groupby(['class', 'sex'])
```

```python
df_grouped.mean()
```

```python
df_grouped.fare.std()
```

```python
df_grouped.mean().xs('male', level='sex')
```

멀티 인덱스 인덱싱  
그룹들 중 성별만 선택하여 female만 값을 지정하여 인덱싱

```python
df_grouped.mean().xs('female', level='sex')
```

평균이 0.5보다 큰 그룹만 필터링

```python
df_grouped.filter(lambda x: x.survived.mean() > 0.5)
```

```python
df_grouped.apply(lambda x: x.survived.mean() > 0.5)
```

```python
df_grouped.mean().loc['Third', 'female']
```
